```python
def extract_year(x):
    x = str(x).strip()
    
    # Case 1: already full year (e.g., 2014)
    if x.isdigit():
        return int(x)
    
    # Case 2: format like "Jul-17"
    else:
        year = int(x.split('-')[1])
        
        # convert 2-digit → 4-digit
        if year <= 25:   # adjust threshold if needed
            return 2000 + year
        else:
            return 1900 + year

df['year'] = df['registration_year'].apply(extract_year)
```
